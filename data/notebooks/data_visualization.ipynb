```python
data = df.drop(columns=["Unnamed: 32", "id", "diagnosis"])
x = data
labels = df.diagnosis 
standarized_data = (data - data.mean()) / (data.std())
```

```python
ax = seaborn.countplot(labels)
B, M = labels.value_counts()
plt.xlabel('Diagnoza')
plt.ylabel('Ilość przykładów')
plt.savefig("/Users/karola/PycharmProjects/breast_cancer/data/plots/count_plot.png", bbox_inches='tight', dpi=500)
# print("Number of benign:", B)
# print("Number of malignant:", M)
# print("Percentage of malignant in whole dataset: {}%".format(np.round(100*M/(B+M),2)))
```

```python
data_visualization = pd.concat([labels,standarized_data.iloc[:,0:10]],axis=1)
data_visualization = data_visualization.rename(index=str, columns={"diagnosis": "Diagnoza"})
plt.figure(figsize=(10,10))
data = pd.melt(data_visualization,id_vars="Diagnoza",var_name="features",value_name='value')
# seaborn.violinplot(x="features", y="value", hue="diagnosis", data=data,split=True, inner="quart")
plt.xticks(rotation=90)
seaborn.boxplot(x='features', y='value', hue='Diagnoza', data=data)
plt.xlabel('Cechy')
plt.ylabel('Wartość')
plt.savefig("/Users/karola/PycharmProjects/breast_cancer/data/plots/box_plot1.png", bbox_inches='tight', dpi=500)
```
